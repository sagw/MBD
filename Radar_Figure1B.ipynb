```r
require(data.table)
require(ggplot2)
require(RColorBrewer)
require(tidyr)
options(repr.plot.width=20, repr.plot.height=15)
```

```r
all=read.csv("~/Documents/Lenfest/Plot_CSV_v2/Fig1Bdata.csv")
```

```r
all$Stat
```

```r
all$Network=factor(all$Network, levels=c("Important Areas", "Representativity", "Connectivity", "Replication", "Viability", "Adequacy"))
```

```r
all$Stat=factor(all$Stat, levels=c("BIAs", "IBAs", "Mangroves", "Seagrasses", "Tropical Corals", "Cold Water Corals", "Climate", "Coverage", "Adequately sized PAs"))
```

```r
all
```

```r
ggplot(all) +
  geom_col(aes(x = Stat, y = Percent, fill = PA),
    position = "identity",
    show.legend = TRUE) +
scale_fill_manual(values=c("goldenrod", "cyan4"))+
  scale_y_continuous(
    limits = c(-15, 100),
    expand = c(0, 0),
    breaks = c(0, 25, 50, 75)
  ) + 
  theme_bw()+
 theme(legend.position="right",  
          panel.grid.major.x = element_blank(), 
          panel.grid = element_line(color="black"),
          axis.ticks.x=element_line(color="black"), 
          axis.line=element_line(color="black"), axis.text=element_text(color="black"),
          text = element_text(size=15))+
  # Make it circular!
  coord_polar()
```

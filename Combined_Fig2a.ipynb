```r
require(data.table)
require(ggplot2)
require(tidyr)
require(RColorBrewer)
options(repr.plot.width=20, repr.plot.height=15)
```

```r
habitats=read.csv("~/Documents/Lenfest/Plot_CSV_v2/habitats.csv", row.names=1)
species=read.csv("~/Documents/Lenfest/Plot_CSV_v2/species.csv", row.names=1)
Importantareas=read.csv("~/Documents/Lenfest/Plot_CSV_v2/Importantareas.csv", row.names=1)
```

```r
colnames(species)=colnames(habitats)
colnames(Importantareas)=colnames(habitats)
```

```r
Importantareas
```

```r
all=rbind(habitats, species, Importantareas)
```

```r
ggplot(all, aes(y=Habitat, x=percent, fill=PA, color=PA)) +
scale_fill_manual(values=c("goldenrod", "cyan4"))+
scale_color_manual(values=c("goldenrod", "cyan4"))+
geom_boxplot(position=position_dodge(0.2), alpha=0.75 )+
geom_point(aes(fill=PA),shape=21,position=position_jitterdodge(dodge.width=0, jitter.height = .5, jitter.width = .5), size=5)+
theme_bw()+
facet_grid(.~Region, scales="free", space="free",    labeller = labeller(Region = label_wrap_gen(10)))+

labs(x="Percent found in PA")+
theme(legend.position="right",  
          panel.grid.major.x = element_blank(), 
          panel.grid.minor.x = element_blank(),

          axis.ticks.x=element_blank(), 
          axis.line=element_line(color="black"), axis.text=element_text(color="black"),
          text = element_text(size=25),
      strip.background = element_blank())+
xlim(0,100)+
ylab("Component")
```

```r
all[all$Region==c("Alaska", "Pacific Islands"),]
```

```r
all$Habitat=factor(all$Habitat, levels=c("Species","BIA", "IBA", "Mangrove", "Seagrass", "CR", "CWC", "Redlist", "Nemesis"))
```

```r
ggplot(all[all$Region=="Pacific Islands"|all$Region=="Alaska",], aes(y=(Habitat), x=percent, fill=PA, color=PA)) +
scale_fill_manual(values=c("goldenrod", "cyan4"))+
scale_color_manual(values=c("goldenrod", "cyan4"))+
geom_boxplot(position=position_dodge(0.2),alpha=0.7, lwd=1.5 )+
#geom_point(aes(fill=PA),shape=21,position=position_jitterdodge(dodge.width=0, jitter.height = .5, jitter.width = .5), size=5)+
theme_bw()+
facet_grid(.~Region, scales="free", space="free",    labeller = labeller(Region = label_wrap_gen(10)))+
scale_y_discrete(limits = rev(levels(all$Habitat)))+
labs(x="Percent found in PA")+
theme(legend.position="right",  
          panel.grid.major.x = element_blank(), 
          panel.grid.minor.x = element_blank(),

          axis.ticks.x=element_blank(), 
          axis.line=element_line(color="black"), axis.text=element_text(color="black"),
          text = element_text(size=25),
            strip.background = element_blank())+

xlim(0,100)+
ylab("Component")
```
